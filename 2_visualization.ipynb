```python
#data analysis
import numpy as np
import pandas as pd

#visualization
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
df = pd.read_csv("1_cleaned.csv")
```

```python
df.head()
```

### __Distribution of Price__

```python
df.groupby("District")["Price"].describe().T
```

```python
plt.figure(figsize=(16,9))
ax = sns.kdeplot(data=df,x="m2_net",shade=True,hue="Building_Age",palette="bright");
```

### __Relationship Between District and Price__

```python
plt.figure(figsize=(48,9))
ax = sns.barplot(data=df,x="District",y="Price");
```

### __Relationship Between Gross and Net Area__

```python
sns.lmplot(data=df,x="m2_gross",y="m2_net",height = 5,aspect = 2);
```

### __Gross area and Price__

```python
sns.lmplot(data=df,x="m2_gross",y="Price",height = 5,aspect = 2);
```

### __Net Area and Price__

```python
sns.lmplot(data=df,x="m2_net",y="Price",height = 5,aspect = 2);
```

### __Relationship Between Number of Rooms and Net Area__

```python
plt.figure(figsize=(21,9))
ax = sns.barplot(data=df,x="Number_of_rooms",y="m2_net");
```

### __Relationship Between Age of Building and Price__

```python
plt.figure(figsize=(21,9))
ax = sns.catplot(data=df,x="Building_Age",y="Price",height=5,aspect=3);
```

### __Floor location and Price__

```python
plt.figure(figsize=(24,9))
ax = sns.catplot(data=df,x="Floor_location",y="Price",height=5,aspect=4);
```

### __Heating type and Price__

```python
plt.figure(figsize=(26,9))
palette = sns.color_palette("hls",8)
ax = sns.barplot(data=df,x="Heating",y="Price",palette = palette);
```

### __Number of bathrooms and net area__

```python
plt.figure(figsize=(16,8))
ax = sns.boxplot(data=df,x="Number_of_bathrooms",y="m2_net");
```
